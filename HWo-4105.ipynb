```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D3.csv')
df.head()
m=len(df)
m
```

```python
X_1 = df.values[:,0]
X_2 = df.values[:,1]
X_3 = df.values[:,2]
Y   = df.values[:,3]
m=len(Y)
print('X_1 = ',X_1[:99])
print('X_2 = ',X_2[:99])
print('X_3 = ',X_3[:99])
print('Y = ',Y[:99])
print('M = ',m)
```

```python
X_0 = np.ones((m, 1))
X_1 = X_1.reshape(m, 1)
X_2 = X_2.reshape(m, 1)
X_3 = X_3.reshape(m, 1)
```

```python
def computeCost(x, y, theta):
    predictions = x.dot(theta)
    errors = np.subtract(predictions, y)
    sqrErrors = np.square(errors)
    J = 1/(2 * m) * np.sum(sqrErrors)
    
    return J
```

```python
def gradientDescent(x, y, theta, alpha, iterations):
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        predictions = x.dot(theta)
        errors = np.subtract(predictions, y)
        sum_delta = (alpha / m) * x.transpose().dot(errors)
        theta = theta - sum_delta;
        cost_history[i] = computeCost(x, y, theta)
        
    return theta, cost_history
```

```python
iterations = 1500
alpha = 0.1
x = np.hstack((X_0, X_1))
theta = np.zeros(2)

theta, cost_history = gradientDescent(x, Y, theta, alpha, iterations)

print('Final value of theta =',theta)
print('cost_history =',cost_history)
```

```python
plt.scatter(X_1, Y, color = 'red', marker = "+", label = 'Training Data')
plt.grid()
plt.rcParams["figure.figsize"]=(10,6)
plt.title('X_1 Scatter')
```

```python
plt.scatter(X_1, Y, color = 'red', marker = "+", label = 'Training Data')
plt.plot(X_1, x.dot(theta), color = 'green', label = 'Linear Regression')
plt.title('X_1 Linear Regression')
plt.grid()
plt.legend()
```

```python
iteration_array = np.arange(1, iterations + 1)

plt.plot(iteration_array, cost_history, color= 'blue', label = 'Cost')
plt.title('Covergence of Gradient Descent for X_1')
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.grid()
```
